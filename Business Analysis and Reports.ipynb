```python
df1['Targeted Sales of Products']= df1.loc[:, 'Jan': 'Dec'].sum(axis=1)
df1
```

```python
products_df= df1[['Product', 'Year', 'Targeted Sales of Products']]
products_df
```

```python
df_merge= products_df.merge(df4, left_on='Product', right_on= 'Product ID' )
df_merge= df_merge.merge(df3, on= 'Material ID')
df_merge.head()
```

```python
columns= ['Product', 'Year', 'Material ID', 'Material Name_x', 'Unit of Measure']
df_merge.drop(columns= columns, inplace= True)
```

```python
columns= ['Product ID', 'Material Name_y','Targeted Sales of Products','Cost per unit', 'Quantity to make 1 unit of Product']
df= df_merge.reindex(columns=columns)
df=df.rename(columns= {'Material Name_y': 'Material Name'})
df=df.sort_values('Product ID', ignore_index=True)
df.head()
```

```python
df['Total Cost for Raw Materials'] = df['Cost per unit'] * df['Targeted Sales of Products'] * df['Quantity to make 1 unit of Product']
df['Total Quantities Required']= df['Quantity to make 1 unit of Product'] * df['Targeted Sales of Products']
df.head()
```

### First table to be included in report

```python
columns= ['Product ID', 'Material Name','Targeted Sales of Products', 'Total Cost for Raw Materials', 'Total Quantities Required']
total_costs_df= df[columns]
total_costs_df= total_costs_df.rename(columns={'Product ID': 'ProductId', 'Material Name': 'MaterialName',
                                              'Targeted Sales of Products': 'Targeted_sales_of_products',
                                              'Total Cost for Raw Materials': 'Total_cost_for_raw_materials',
                                              'Total Quantities Required': 'Total_quantities_required'})
total_costs_df['Total_cost_for_raw_materials']=total_costs_df['Total_cost_for_raw_materials'].round(2)
total_costs_df['Total_quantities_required']=total_costs_df['Total_quantities_required'].round(2)

total_costs_df
```

### Second table to be included in Report

```python
agg_cost= total_costs_df[['ProductId', 'Total_cost_for_raw_materials']]

agg_cost= agg_cost.groupby('ProductId')['Total_cost_for_raw_materials'].sum()
```

```python
agg_cost_df=pd.DataFrame(agg_cost)
agg_cost_df.reset_index(inplace=True)
```

```python
agg_cost_df= agg_cost_df.rename(columns={'Product ID': 'ProductId', 
                                         'Total Cost for Raw Materials': 'Total_cost_for_raw_materials'})
agg_cost_df
```

### Third task

```python
profit_30_df= df1.merge(agg_cost_df, left_on= 'Product', right_on= 'ProductId')
profit_30_df.drop(columns=['ProductId','Year'], inplace= True)
profit_30_df['Profit 30% on Raw Materials']=profit_30_df['Total_cost_for_raw_materials'] + (30/100 * profit_30_df['Total_cost_for_raw_materials'])
```

```python
profit_30_df['Selling Price for Each Product']= profit_30_df['Profit 30% on Raw Materials'] / profit_30_df['Targeted Sales of Products']
profit_30_df
```

```python
months= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul','Aug', 'Sep','Oct','Nov','Dec']
profit_30_df[months]= profit_30_df[months].multiply(profit_30_df['Selling Price for Each Product'], axis=0)
profit_30_df
```

### Generating our pdf reports

```python
#creating a function to obtain attributes from our dataframe so we can pass into a table in our pdf report
def data_to_pdf(pdf, df):
    table_cell_width= 40
    table_cell_height=6
    pdf.set_font('Arial', 'B', 8)
    cols=df.columns
    
    for col in cols:
        pdf.cell(table_cell_width, table_cell_height, col, align='C', border=1)
    pdf.ln(table_cell_height)
    pdf.set_font('Arial', '', 10)
    for row in df.itertuples():
        for col in cols:
            value= str(getattr(row,col))
            pdf.cell(table_cell_width, table_cell_height, value, align='C', border =1)
        pdf.ln(table_cell_height)
```

```python
#generating our pdf report
class PDF(FPDF):
    def header(self):
        #logo
        self.image('C:/Users/user/Desktop/SmatData.jpg', 10,8,25)
        #font
        self.set_font('Arial', 'B', 20)
        #padding
        self.cell(80)
        #title
        self.cell(30,10,'Report', border=True, ln=1, align='C')
        #linebreak
        self.ln(5)

pdf= PDF('P','mm','Letter')
pdf.add_page()
pdf.set_font('Arial', 'B', 16)

pdf.cell(40,10,'Total Cost of Materials to Produce Targeted Sales')
pdf.ln(10)
data_to_pdf(pdf, total_costs_df)


pdf.ln(10)

pdf.set_font('Arial', 'B', 16)
pdf.cell(40,10, 'Aggregate Cost of Raw Materials')
pdf.ln(10)
data_to_pdf(pdf, agg_cost_df)


pdf.output('Smatdata Report.pdf', 'F')
```
